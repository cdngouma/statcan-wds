```python
import pandas as pd
from statcan_wds import get_table_data
import json
import yaml
```

```python
def load_cfg(path, fmt="yaml"):
    if fmt == "yaml" or fmt == "yml":
        with open(path, "r", encoding="utf-8") as f:
            return yaml.safe_load(f)
    elif fmt == "json":
        with open(path, "r", encoding="utf-8") as f:
            return json.load(f)
```

### Consumer Price Index, monthly, seasonally adjusted [pid: 18100006]

```python
cpi_cfg = load_cfg(path="examples/cpi_adjusted.yaml")
```

```python
cpi_df = get_table_data(
    pid=cpi_cfg["pid"],
    query_spec=cpi_cfg.get("query", None),
    ref_start=cpi_cfg.get("ref_start", None),
    ref_end=cpi_cfg.get("ref_end", None)
)
```
